```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Criando dados fictícios para rescisões
# Cria um range de datas com frequência diária 'D'
datas = pd.date_range(start='2024-01-01', end='2024-10-31', freq='D')

# Criando valores aleatórios de rescisão em um intervalo de valores entre 15.000 e 30.000
valores_rescisao = np.random.randint(15000, 30001, size=len(datas))

# Trasformando em datframe
df = pd.DataFrame({
    'data_rescisao': datas,
    'valor_rescisao': valores_rescisao
})

# Agrupando o custo total com rescisões por mês
resultado = df.groupby(pd.Grouper(key='data_rescisao', freq='ME')).sum()

# Ajustando o índice para exibir apenas mês e ano
resultado.index = resultado.index.strftime('%m-%Y')
print(resultado)
```

```python
# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(resultado.index, resultado['valor_rescisao'], marker='o', linestyle='-', color='b')
plt.title('CUSTO MENSAL COM RESCISÕES DE CONTRATO DE TRABALHO EM 2024')
plt.xlabel('MÊS-ANO')
plt.ylabel('VALOR EM MILHÕES (R$)')
plt.xticks(rotation=0)
plt.grid(True)
# Definindo a escala do eixo Y para mostrar de 350.000 em 350.000
plt.yticks(np.arange(0, resultado['valor_rescisao'].max() + 350000, 350000))
plt.tight_layout()
plt.show()
```
